```python
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
files = glob.glob("../**/*.log", recursive=True)
```

```python
files
```

```python
nomes = [file.split("/")[-2] for file in files]
logs = [file.split("/")[-1].split(".")[0] for file in files]
```

```python
def plot_csv(acc, val_acc, loss, val_loss, nome):
    plt.subplot()
    plt.rcParams['figure.figsize'] = (10.0, 6.0)
    plt.title('Baseline Model Accuracy')
    plt.plot(acc, lw=3, marker="o")
    plt.plot(val_acc, lw=3, marker="o")
    plt.ylabel('Accuracy')
    plt.ylim((0, 1.05))
    plt.xlabel('Epochs')
    plt.legend(['Training Accuracy','Validation Accuracy'], loc=4)
    plt.title(nome)
    plt.savefig("./PLOTS/acc" + "_".join(nome.split(" ")), transparent= False, bbox_inches= 'tight', dpi= 400)

    plt.show()
        
    plt.subplot()
    plt.title('Baseline Model Loss')
    plt.plot(loss, lw=3, marker="o")
    plt.plot(val_loss, lw=3, marker="o")
    plt.ylabel('Loss')
    plt.xlabel('Epochs')
    plt.legend(['Training Loss','Validation Loss'], loc=1)
    if("RANDOM" in nome):
        plt.ylim((0, 10))
    else:
        plt.ylim((0, 10))
    plt.title(nome)
    plt.savefig("./PLOTS/loss" + "_".join(nome.split(" ")),transparent= False, bbox_inches= 'tight', dpi= 400)
    plt.show()
```

```python
for file, nome, log in zip(files, nomes, logs):
    print(file)
    log_csv = pd.read_csv(file)
    nome = " ".join(nome.split("-"))
    nome += " "
    nome += " ".join(log.split("_")[:2])
    plot_csv(log_csv.accuracy, log_csv.val_accuracy, log_csv.loss, log_csv.val_loss, nome)
```
