```python
#Importing Necessary Libraries.
from PIL import Image
import numpy as np
import pandas as pd
import os
import cv2
import matplotlib.pyplot as plt
from glob import glob
%matplotlib inline
```

```python
def showSomeNxN(N, path, page):
    bmps = glob(path)
    fig, axes = plt.subplots(N, N, figsize=(20, 20))
    for i in range(N):
        for j in range(N):
            plt.subplot(N, N, N*i+j+1)
            iar = cv2.imread(bmps[N*i+j+1*page])
            plt.imshow(iar)
            print(iar.shape)
            plt.subplots_adjust(wspace=0, hspace=0)
    plt.show()


showSomeNxN(5, '../EXP1/chest_xray/train/NORMAL/*.jpeg', 1)
showSomeNxN(5, '../EXP1//chest_xray/train/PNEUMONIA/*.jpeg', 1)
```

```python
data=[]
labels=[]
shapesx = []
shapesy = []

NORMAL=os.listdir("../EXP1/chest_xray/train/NORMAL/")
for a in NORMAL:
    try:
        image=cv2.imread("../EXP1/chest_xray/train/NORMAL/"+a)
        shapesx.append(image.shape[0])
        shapesy.append(image.shape[1])

    except AttributeError:
        print(a)

PNEUMONIA=os.listdir("../EXP1/chest_xray/train/PNEUMONIA/")
for b in PNEUMONIA:
    try:
        image=cv2.imread("../EXP1/chest_xray/train/PNEUMONIA/"+b)
        shapesx.append(image.shape[0])
        shapesy.append(image.shape[1])

    except AttributeError:
        print(b)
```

```python
fig = plt.figure(1, figsize=(15, 10))

xaxis = fig.add_subplot(121)
yaxis = fig.add_subplot(122)

xaxis.boxplot(shapesx)
yaxis.boxplot(shapesy)

plt.show()
print(np.median(shapesx))
print(np.median(shapesy))
```

```python
NORMAL_TRAIN_X=[]
NORMAL_TRAIN_Y=[]
NORMAL_TEST_X=[]
NORMAL_TEST_Y=[]

PNEUMONIA_TRAIN_X=[]
PNEUMONIA_TRAIN_Y=[]
PNEUMONIA_TEST_X=[]
PNEUMONIA_TEST_Y=[]

h_input = 512
w_input = 512

def load_append(path, data, labels, label):
    path_list=os.listdir(path)
    for a in path_list:
        try:
            image=cv2.imread(path+a)
            image_from_array = Image.fromarray(image, 'RGB')
            size_image = image_from_array.resize((w_input, h_input))
            data.append(np.array(size_image))
            labels.append(label)
        except AttributeError:
            print(a)
```

```python
load_append("../EXP1/chest_xray/train/NORMAL/", NORMAL_TRAIN_X, NORMAL_TRAIN_Y, 0)
NORMAL_TRAIN_x=np.array(NORMAL_TRAIN_X)
np.save("./NORMAL_TRAIN_X",NORMAL_TRAIN_x)

NORMAL_TRAIN_Y=np.array(NORMAL_TRAIN_Y)
np.save("./NORMAL_TRAIN_Y",NORMAL_TRAIN_Y)
```
